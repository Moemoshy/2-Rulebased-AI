```python
import pandas as pd

# Load the dataset
df = pd.read_excel("company_access_data.xlsx")

# Build rules dictionary
rules = {}
for index, row in df.iterrows():
    # Convert everything to string first, then normalize
    emp_id = str(row['EmployeeID']).strip().lower()
    password = str(row['Password']).strip().lower()

    rules[emp_id] = {
        'Password': password,
        'Role': str(row['Role']),
        'Name': str(row['Name'])
    }

# Backward chaining check
def check_access(user_id, user_pass):
    user_id_norm = str(user_id).strip().lower()
    user_pass_norm = str(user_pass).strip().lower()

    if user_id_norm not in rules:
        return False, "Invalid Employee ID."

    if user_pass_norm != rules[user_id_norm]['Password']:
        return False, "Incorrect password."

    return True, rules[user_id_norm]['Name']

# Authentication loop with 2 attempts
def authenticate_user():
    attempts = 2
    while attempts > 0:
        print("\n=== ACCESS CONTROL LOGIN ===")
        user_id = input("Enter Employee ID: ")
        user_pass = input("Enter Password: ")

        access_granted, info = check_access(user_id, user_pass)

        if access_granted:
            print(f"\nACCESS GRANTED. Hello, {info}!")
            return
        else:
            print(f"Access Denied: {info}")
            attempts -= 1

    print("\n‚ùå You have been blocked after 2 failed attempts.")

# Run system
authenticate_user()
```
