```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier

# Load dataset
df = pd.read_excel("company_access_data.xlsx")

# Encode categorical features
le_role = LabelEncoder()
df['RoleEncoded'] = le_role.fit_transform(df['Role'])

le_id = LabelEncoder()
df['IDEncoded'] = le_id.fit_transform(df['EmployeeID'])

le_pass = LabelEncoder()
df['PassEncoded'] = le_pass.fit_transform(df['Password'])

# Features + target
X = df[['IDEncoded', 'PassEncoded', 'RoleEncoded']]
y = df['EmployeeID']

# Train model
model = DecisionTreeClassifier()
model.fit(X, y)

def authenticate_user():
    attempts = 2

    while attempts > 0:
        print("\n=== ACCESS CONTROL LOGIN ===")
        user_id = input("Enter Employee ID: ").strip()
        user_pass = input("Enter Password: ").strip()

        # Check if ID exists first
        if user_id not in df['EmployeeID'].values:
            print("Invalid ID.")
            attempts -= 1
            continue

        # Try encoding password — invalid passwords will fail
        try:
            encoded_id = le_id.transform([user_id])[0]
            encoded_pass = le_pass.transform([user_pass])[0]
        except:
            print("Incorrect password.")
            attempts -= 1
            continue

        # Get role for encoding
        user_role = df[df['EmployeeID'] == user_id]['Role'].values[0]
        encoded_role = le_role.transform([user_role])[0]

        # Predict
        prediction = model.predict([[encoded_id, encoded_pass, encoded_role]])[0]

        if prediction == user_id:
            user_name = df[df['EmployeeID'] == user_id]['Name'].values[0]
            print(f"\nACCESS GRANTED. Hello, {user_name}!")
            return
        else:
            print("Access Denied.")
            attempts -= 1

    print("\n ⨉ You have been blocked after 2 failed attempts.")

# Run system
authenticate_user()
```
